```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import LinearRegression
```

```python
dados = pd.read_csv('Wine.csv')
dados = dados[["Alcohol", "Magnesium", "Customer_Segment"]]

dados.head()
```

```python
x = dados.drop(columns="Customer_Segment")
y = dados.Customer_Segment

np.random.seed(42)
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.2)
```

```python
class Processt(BaseEstimator, TransformerMixin):
    def __init__(self, feature_name):
        print("\n>>>>init() called\n")
        self.feature_name = feature_name
        
    def fit(self, x, y=None):
        print("\n>>>>fit() called\n")
        return self
        
    def transform(self, x, y=None):
        print("\n>>>>transform() called\n")
        x_ = x.copy()
        x_["nova"] = x_[self.feature_name] ** 2
        print(x_)
        return x_
```

```python
process = Processt("Magnesium")
process.fit(treino_x, treino_y)
process.transform(treino_x, treino_y)
```

```python
pipeline = Pipeline(steps=[('teste', Processt("Magnesium")),
                           ('model', LinearRegression())
                          ])

pipeline.fit(treino_x, treino_y)
pipeline.predict(teste_x)
```
